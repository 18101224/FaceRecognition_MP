```python
import torch
from torch import optim

weight = torch.randn((7,512),requires_grad=True)
opt = optim.Adam([weight],lr=0.001)
for i in range(100000):
    opt.zero_grad()
    w = torch.nn.functional.normalize(weight,dim=1,p=2)
    sims = torch.triu(w@w.T)
    sims.fill_diagonal_(0)
    std = torch.std(sims)
    loss = torch.sum(sims,dim=[0,1],keepdim=False)+std
    loss.backward()
    opt.step()
```
